### There are two common packages to convert PDF files to Excel; PDFTables is one of of the two

````python
# !pip install pdftables-api
````

````python
# importing the pdftables_api package
import pdftables_api
````

````python
# the path to the input pdfs
first_pdf = r"../pdf_files/2016-financial-statements-en.pdf"
second_pdf = r"../pdf_files/ab-group-fs_2020-(english).pdf"

# the output file path
excel_path = r"../excel_files/"
````

````python
# initialization, the api_key obtained after you create an account on the PDFTables.com website
c = pdftables_api.Client(api_key='hncf7esx1ldx')
````

````python
# let's try to convert the first pdf into excel-file (.xlsx)
c.xlsx(first_pdf, f"{excel_path}first") 

#replace c.xlsx with c.csv to convert to CSV or to c.xml to convert to XMLor to c.html to convert to HTML
````

### This Python library needs to interact with the ```PDFTables.com``` API. so it's very limited 

### Notice the coming error is APIException: Usage limit exceeded

````python
# let's make another try with the second file this time.
c.xlsx(second_pdf, f"{excel_path}second.xlsx")
````

### now let's show the resulted excel file

````python
import pandas as pd
excel_dict = pd.read_excel(f"{excel_path}.xlsx", sheet_name=None)
````
